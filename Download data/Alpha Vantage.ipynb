```python
# Import necessary libraries
import pandas as pd
from alpha_vantage.timeseries import TimeSeries

# Alpha Vantage API key
api_key = 'N5OGBA3J7I61X40X'

# Create the TimeSeries client
ts = TimeSeries(key=api_key, output_format='pandas')

# Download full daily data 
data, meta_data = ts.get_daily(symbol='SPY', outputsize='full')

# Keep only the close price and rename the column
data = data[['4. close']].copy()
data.columns = ['close']

# Convert the index to datetime and sort it
data.index = pd.to_datetime(data.index)
data = data.sort_index()

# Desired date range
start_date = '2015-01-01'
end_date   = '2025-01-01'
filtered = data.loc[start_date:end_date]

# Save data to CSV
filtered.to_csv('data.csv')

# Verification prints
print(f"Requested period: {start_date} → {end_date}")
print(f"Saved rows:       {len(filtered)}")
print("First 5 dates:\n", filtered.head(), "\n")
print("Last  5 dates:\n", filtered.tail())
```
